```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression

df = pd.read_csv("data/student_info.csv")
df = df.dropna()

plt.figure(figsize=(14, 8))
with sns.axes_style("whitegrid"):
    sns.scatterplot(data=df,
                    x="study_hours",
                    y="student_marks",
                    hue="student_marks",
                    size="student_marks")
    
    model = LinearRegression()
    model.fit(df[['study_hours']], df['student_marks'])
    
    X = df['study_hours'].values.reshape(-1, 1)
    y = model.predict(X)
    
    plt.plot(X, y, color='red', label='Linear Regression Line')
    plt.xlabel('Study Hours')
    plt.ylabel('Student Marks')
    plt.title('Scatter Plot with Linear Regression Line')
    plt.legend()
    plt.show()

print(f" model intercept: {model.intercept_}\n coef:{model.coef_}\n and score is {model.score(df[['study_hours']], df['student_marks'])}")
```

```python
sns.regplot(data=df, 
            x='study_hours',
            y='student_marks')

plt.figure(figsize=(8,4), dpi=200)
with sns.axes_style("whitegrid"):
  sns.regplot(data=df, 
            x='study_hours', 
            y='student_marks',
            scatter_kws = {'alpha': 0.2},
            line_kws = {'color': 'black'})
```
